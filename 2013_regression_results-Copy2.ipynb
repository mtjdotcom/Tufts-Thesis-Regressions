## 2013 Data: Regression Results copy

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.read_csv('data/refined_data/2013_reg_data_mod.csv', index_col=0)
df.head()
```

```python
df.corr()
```

```python
df['log_btc_dl'] = np.log(df.btc_dl)
df['log_remitt_in_13'] = np.log(df.remitt_in_13)
df['log_gdp_13'] = np.log(df.gdp_13)
df['log_mig_to_OECD'] = np.log(df.mig_to_OECD)
```

```python
# set the dependent variable

y = df.log_btc_dl
```

```python
# regression 1: 
# depedent variable = log of bitcoin downloads 11/1/12-11/1/13. A cutoff of November 2013 is used
# to control for the run up in BTC/USD exchange rate in Nov 2013. 

# independet variable = log of remittances to a country

x = df.log_remitt_in_13
x = sm.add_constant(x)
est1 = sm.OLS(y, x, missing='drop')
est1 = est1.fit()
print est1.summary()
```

```python
# regression 2: 
# depedent variable = log of bitcoin downloads (11/1/12-11/1/13)

# independet variables = log of remittances to a country, log of GDP

x1 = df[['log_gdp_13', 'log_remitt_in_13']]
x1 = sm.add_constant(x1)
est2 = sm.OLS(y, x1, missing='drop')
est2 = est2.fit()
print est2.summary()
```

```python
# regression 3: 
# depedent variable = log of bitcoin downloads (11/1/12-11/1/13)

# independet variables = log of remittances to a country, log of GDP, internet usesrs per 100 people, mobile phone
# subscriptions per 100

x2 = df[['log_gdp_13', 'log_remitt_in_13', 'int_users_per100', 'mobile_phone_13']]
x2 = sm.add_constant(x2)
est3 = sm.OLS(y, x2, missing='drop')
est3 = est3.fit()
print est3.summary()
```

```python
# regression 4: 
# depedent variable = log of bitcoin downloads (11/1/12-11/1/13)

# independet variable = log of remittances to a country, log of GDP, internet usesrs per 100 people, mobile phone
# subscriptions per 100, percentage of economy that is considered to be black market, and rate of 2013 inflation

x3 = df[['log_gdp_13', 'log_remitt_in_13', 'inflation_13', 'blk_market_08', 'mobile_phone_13', 'int_users_per100']]
x3 = sm.add_constant(x3)
est4 = sm.OLS(y, x3, missing='drop')
est4 = est4.fit()
print est4.summary()
```

```python
# regression 5: 
# depedent variable = log of bitcoin downloads (11/1/12-11/1/13)

# independet variable = log of remittances to a country, log of GDP, internet usesrs per 100 people, mobile phone
# subscriptions per 100, percentage of economy that is considered to be black market, and rate of 2013 inflation
# rate of migration to OECD countries

x4 = df[['log_gdp_13', 'log_remitt_in_13', 'inflation_13', 'blk_market_08', 'mobile_phone_13', 'log_mig_to_OECD', 'int_users_per100']]
x4 = sm.add_constant(x4)
est5 = sm.OLS(y, x4, missing='drop')
est5 = est5.fit()
print est5.summary()
```

```python
# regression 6: 
# depedent variable = log of bitcoin downloads (11/1/12-11/1/13)

# independet variable = log of remittances to a country, log of GDP, internet usesrs per 100 people, mobile phone
# subscriptions per 100, and rate of 2013 inflation

x7 = df[['log_gdp_13', 'log_remitt_in_13', 'inflation_13', 'mobile_phone_13', 'int_users_per100']]
x7 = sm.add_constant(x7)
est8 = sm.OLS(y, x7, missing='drop')
est8 = est8.fit()
print est8.summary()
```
